```python
import numpy as np
print(np.__version__)
```

```python
mylist = [1,2,3,4]
mylist = mylist*2 #Normal python list and operation
print(mylist)
```

## Normal Python simply repeats the list when multiplied by 2 but NumPy is superior and it can be used to make a numpy array by using np.array().

```python
array = np.array([1,2,3,5])
print(array)
```

To check whether our array is numpy array or not.

```python
print(type(array))
```

In order to multiply all our elements with two-->


```python
array = array * 2
print(array)
```

# Multidimensional Arrays 
A multidimensional array is an array that has more than one dimension (axis).
In NumPy, this means arrays like 2D, 3D, or higher, instead of a simple 1D list.

| Type | Example         | Shape     |
| ---- | --------------- | --------- |
| 1D   | `[1, 2, 3]`     | `(3,)`    |
| 2D   | `[[1,2],[3,4]]` | `(2,2)`   |
| 3D   | `[[[...]]]`     | `(x,y,z)` |



```python
array = np.array('A')
print(array.ndim) #prints the number of dimension of our array
```

```python
array = np.array(['A','B','C'])
print(array.ndim) #prints the number of dimension of our array
```

```python
array = np.array([[['A','B','C'],['D','E','F'],['G','H','I']]])
print(array.ndim) #prints the number of dimension of our array
```

We can also print the shape of the array created using numpy.

```python
print(array.shape) #depth, number of rows and number of columns
```
